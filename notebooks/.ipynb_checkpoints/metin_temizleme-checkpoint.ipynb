```python
import re
import json
```

```python
#dosyayı oku
with open("../data/processed/turkce.txt", "r", encoding="utf-8") as f: #dosyayı okuma modunda 'r' ile açar.
    text= f.read()

#soruları regex ile böl
#bu desen "1.", "2."... gibi numaralı soruları yakalar
soru_deseni= r"\d+\.\s*(.*?)(?=(\d+\.\s)|$)"
#/d+ -> Bir veya daha fazla rakamı bulur(örneğin 1,2...) 
#/. -> Ardından gelen noktayı bulur
#/s* -> Noktadan sonra gelen boşlukları yok sayar.
#(.*?) ->  Soru bloğunun kendisini mümükün olan en kısa şekilde yazar.
#(?=(\d+\.\s)|$) -> Bir sonraki soru numarasına veya metin sonucuna kadar olan ksımı alır.
sorular= re.findall(soru_deseni, text, flags= re.S)

data= [] #boş liste oluştur
```

```python
# Her soru bloğunu işle
for i, (blok, _) in enumerate(sorular, start=1): #enumerate her soru bloğuna bir sıra numarası(id) verir.
    #blok, _) -> findall() çift döndürdüğü için sadece ilk kısmı (blok) alıyoruz.  "_" kullanılmayan değişken demek
    
    # Seçenekleri ayır
    secenekler = re.findall(
    r"([A-E]\)\s*(.*?)(?=(?:[A-E]\)|Cevap:|$)))",
    blok,
    flags=re.S
)

    #([A-D])-> A,B,C,D,E harflerinden biirni alır
    #/) ->Parantez işaretini arar (örneğin A) ) 
    #/s* -> Boşlukları atlar 
    #(.*?) -> bu seçeneğin metnini alır
    #(?=(?:[A-D]\)|→|$)) -> Bir sonraki seçenek, -> Cevap:" veya metin sonuna kadar olan kısmı kapsar.
    sec_dict = {harf: metin.strip() for harf, metin in secenekler} #seçenekelri sözlüğe dönüştürme. strip fonk. baştaki/sondaki boşluları siler.

    # Cevabı bul
    cevap = re.search(r"→\s*Cevap:\s*([A-E])", blok) # Metin içinde "cevap A" kalıbını arar.
    cevap = cevap.group(1) if cevap else None #Parantez içindeki eşleşmeyi (harfi) döndürür. Cevap bulamazsa none atanır.

    # Soru metnini (seçenekler öncesi kısmı) bul
    soru_metni = re.split(r"[A-E]\)", blok)[0].strip() #blok[0]-> ilk kısmı, yani sorunun kendisini alır. strip() gereksiz boşlukları siler

    data.append({ #her şeyi tek bir kayıt haline getirme
        "id": i,
        "soru": soru_metni,
        "secenekler": sec_dict,
        "cevap": cevap
    })
#JSON olarak kaydet
output_path = "../data/processed/turkce_questions.json"
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4) #ensure_ascii -> türkçe karakterler bozulmasın diye
    #indnent=4-> JSON dosysını okunabilir halde, girintili kaydeder.

print(f"✅ {len(data)} soru işlendi ve kaydedildi: {output_path}")
```
