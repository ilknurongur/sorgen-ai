```python
from PyPDF2 import PdfReader
import os
```

```python
pdf_path= "../data/raw/turkce.pdf"
output_txt ="../data/processed/turkce.txt"
```

```python
reader =PdfReader(pdf_path) #burada pdf'i açtın
text= ""
for page in reader.pages: #döngü ile her sayfayı tek tek okuruz.
    text +=page.extract_text()+ "/n" #o sayfadaki metni (yazıyı) çıkartır ve text değişkenine ekleriz. Her sayfanın sonun /n ile bir satır boşluk bırakır.
    #sayfalar birbirine karışmaz.

#kaydetmek için
os.makedirs("data/processed", exist_ok =True) #belirtilen dizini oluşturur. eğer bir klasör varsa hata vermez, geçer
with open(output_txt, "w", encoding="utf-8" ) as f: #yeni bir dosya aç varsa üzeirne yazar. 'w'->yazma modudur.
    f.write(text) #çektiğimiz tüm metni .txt dosyasına yazar. with bloğu bittiğinde dosya otomatik kapanır.

print("PDF metne dönüştürüldü ve kaydedildi", output_txt)
```
